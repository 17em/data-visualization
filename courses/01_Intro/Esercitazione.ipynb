#### Massimo di due numeri

```
def massimo_due_numeri(a,b):
    if a==b:
        return print('i due numeri sono identici')
    elif a>b:
        return print('il numero '+str(a)+' è maggiore di '+ str(b))
    else:
        return print('il numero '+str(b)+' è maggiore di '+ str(a))
```

```
massimo_due_numeri(20,-21)
```

#### Numero pari o dispari

```
numero=float(input('inserisci un numero a tua scelta'))

resto=numero % 2
if resto>0:
    print('il num è dispari')
else:
    print('il num è pari')
```

```
def pari_dispari(numero):
    resto= numero % 2
    if resto>0:
        return print('il num è dispari')
    else:
        return print('il num è pari')
```

#### Cerchio -area+circ.

```
raggio=float(input("inserisci il valore (positivo) del raggio in mm"))
pi=3.14
if raggio<=0:
    print("inserisciun numero positivo del raggio")
else:
    area = raggio**2*pi
    circonferenza = 2*pi*raggio
    print("l'area di questo cerchio è pari a: "+ str(area)+"mm²")
    print("la circonferenza di questo cerchio è pari a: "+ str(circonferenza)+"mm")
```

```
def formule_cerchio(raggio):
    pi=3.14
    if raggio<=0:
        return print("inserisciun numero positivo del raggio")
    else:
        area = raggio**2*pi
        circonferenza = 2*pi*raggio
        return print("l'area di questo cerchio è pari a: "+ str(area)+"mm²\n" 
                     "la circonferenza di questo cerchio è pari a: "+ str(circonferenza)+" mm")
```

```
formule_cerchio(5)
```
