la retta di semplice regressione lineare y= b0 + b1*x + e

```
import numpy as np
generate_random = np.random.RandomState(667)
x = 10 * generate_random.rand(100)
```

```
y = 3 * x + np.random.randn(100)
```

```
plt.figure(figsize = (10, 8))
plt.scatter(x, y);
```

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression(fit_intercept=True)
model
```

```
x
```

```
x.ndim
```

```
X = x.reshape(-1, 1)
X.shape
```

```
X.ndim
```

```
X
```

```
model.fit(X, y)
```

```
model.coef_
```

```
model.intercept_
```

### Apply trained model to a new data

```
X_fit = np.linspace(-1, 11)
```

```
X_fit = X_fit.reshape(-1,1)
```

```
y_fit = model.predict(X_fit)
```

### Visualize

```
plt.figure(figsize = (10, 8))

plt.scatter(x, y)
plt.plot(X_fit, y_fit);
```

```
# store feature matrix in "x"
X = df.drop(['class'], axis=1)
# store response vector in "y"
y = df['class']
```

```
X.shape
```

```
y.shape
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```
from IPython.display import Image
Image(url='https://frenzy86.s3.eu-west-2.amazonaws.com/fav/iris_8.png',width=900,height=300)
```

```
#Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=667)
print(y_test)
```

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train) #Training the model
```

```
#Train the model with Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train) #Training the model
```

```
X_test.shape
```

```
model.predict_proba(X_test)
```

```
#Test the model
y_predict  = model.predict(X_test)
print(y_predict )# printing predictions
```

```
confronto = pd.DataFrame(data = [y_predict,y_test])
confronto = confronto.T
confronto.columns = ['predetti', 'reali_test']
confronto
```

```
length = y_predict.shape[0] # 30 
x = np.linspace(0,length,length)


plt.figure(figsize=(20,15))
plt.plot(x, y_test, label='test')
plt.plot(x, y_predict, label='predetto')
plt.legend(loc=2);
```
